```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import unidecode
import os
```

#### Funciones

```python
def doClean(object):   
    
    # Quitar las expresiones tras la barra /
    #a = re.sub('\/.*', '', object)
    # Quitar las aclaraciones entre paréntesis
    #a = re.sub('\(.*\)', '', object)
    # Quitar acentos 
    a = unidecode.unidecode(object)
    # Quitar aquello que no sean palabras o cosa que se le parezca
    #a = re.sub('\W', ' ', a) 
    # Quitar espacios extra en caso de haber
    a = re.sub('\s+', ' ', a)
    # Pasar el texto a minúsulas
    a = a.lower() 
    a = re.sub(' ', '_', a)
    a = re.sub('-', '_', a)
    #a = a.strip() 
    
    return a
```

```python
#maestro_municipios = pd.read_csv("maestro_municipios.csv", index_col=0,dtype={'cp': object})
maestro_municipios = pd.read_csv("maestro_idiomas.csv", index_col=0,dtype={'postal_code': object})
```

```python
maestro_municipios.head()
```

```python
maestro_municipios['cities_es'] = maestro_municipios['cities_es'].apply(doClean)
maestro_municipios['cities'] = maestro_municipios['cities'].apply(doClean)
maestro_municipios['province_es'] = maestro_municipios['province_es'].apply(doClean)
maestro_municipios['province'] = maestro_municipios['province'].apply(doClean)
maestro_municipios['ccaa_es'] = maestro_municipios['ccaa_es'].apply(doClean)
maestro_municipios['ccaa'] = maestro_municipios['ccaa_es'].apply(doClean)
maestro_municipios['cities_es/province_es'] = maestro_municipios['cities_es/province_es'].apply(doClean)
maestro_municipios['cities_es/province'] = maestro_municipios['cities_es/province'].apply(doClean)
maestro_municipios['cities_es/ccaa_es'] = maestro_municipios['cities_es/ccaa_es'].apply(doClean)
maestro_municipios['cities_es/ccaa'] = maestro_municipios['cities_es/ccaa'].apply(doClean)
maestro_municipios['cities/province_es'] = maestro_municipios['cities/province_es'].apply(doClean)
maestro_municipios['province_es/ccaa_es'] = maestro_municipios['province_es/ccaa_es'].apply(doClean)
maestro_municipios['province_es/ccaa'] = maestro_municipios['province_es/ccaa'].apply(doClean)
maestro_municipios['province/ccaa_es'] = maestro_municipios['province/ccaa_es'].apply(doClean)
maestro_municipios['province/ccaa'] = maestro_municipios['province/ccaa'].apply(doClean)
```
