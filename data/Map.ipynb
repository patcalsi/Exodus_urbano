```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style("white")
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats
```

```python
Combined_data = pd.read_csv('df_kmeans.csv')
Combined_data.head()
```

```python
#Dos lineas de código

colores=['pink','red','green','orange','cyan','yellow','black'] #'black','pink','brown','purple']

fig, axes = plt.subplots(1, figsize=(25,65))
sns.scatterplot(x= Combined_data['longitud'], y=Combined_data['latitud'],palette=colores,hue= Combined_data['cluster'], legend = 'full', ax = axes)

```

```python
#Combined_data['cluster'].value_counts()
```

```python
indice_drop=Combined_data[Combined_data['latitud'] < 0].index
Combined_data.drop(labels=indice_drop, axis=0, inplace=True)
```

```python
colores=['pink','red','green','orange','cyan','yellow','black'] #'black','pink','brown','purple']

fig, axes = plt.subplots(1, figsize=(30,30))
sns.scatterplot(x= Combined_data['longitud'], y=Combined_data['latitud'],alpha=0.9,palette=colores,hue= Combined_data['cluster'].astype('category'),legend = 'full', ax = axes)
```

```python
indice_drop_canarias=Combined_data[Combined_data['longitud'] < -10].index
Combined_data_sin_canarias=Combined_data.drop(labels=indice_drop_canarias, axis=0, inplace=False)
```

```python
fig, axes = plt.subplots(1, figsize=(40,30))
sns.scatterplot(x= Combined_data_sin_canarias['longitud'], y=Combined_data_sin_canarias['latitud'],alpha=1,palette=colores,hue= Combined_data_sin_canarias['cluster'].astype('category'), legend = 'full', ax = axes)
```

```python
filtered = pd.read_csv('df_kmeans_filtrado.csv')
indice_drop=filtered[filtered['latitud'] < 0].index
filtered.drop(labels=indice_drop, axis=0, inplace=True)
```

```python
fig, axes = plt.subplots(1, figsize=(20,20))
sns.scatterplot(x= filtered['longitud'], y=filtered['latitud'],alpha=1,palette=colores,hue= filtered['cluster'].astype('category'), legend = 'full', ax = axes)
```

```python
indice_drop_canarias=filtered[filtered['longitud'] < -10].index
filtered_sin_canarias=filtered.drop(labels=indice_drop_canarias, axis=0, inplace=False)
fig, axes = plt.subplots(1, figsize=(25,20))
sns.scatterplot(x= filtered_sin_canarias['longitud'], y=filtered_sin_canarias['latitud'],alpha=1,palette=colores,hue= filtered_sin_canarias['cluster'].astype('category'), legend = 'full', ax = axes)
```
