```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import unidecode
import os
```

#### Funciones

```python
def doClean(a):   
    
    # Quitar las expresiones tras la barra /
    a = re.sub('\/.*', '', a)
    # Quitar las aclaraciones entre paréntesis
    a = re.sub('\(.*\)', '', a)
    # Quitar acentos 
    a = unidecode.unidecode(a)
    # Quitar aquello que no sean palabras o cosa que se le parezca
    a = re.sub('\W', ' ', a) 
    # Quitar espacios extra en caso de haber
    a = re.sub('\s+', ' ', a)
    # Pasar el texto a minúsulas
    a = a.lower() 
    # Quitar espacio izq
    a = a.lstrip()
    # Quitar espacio derecha
    a = a.rstrip()
    a = re.sub(' ', '_', a)
    #a = a.strip() 
    
    return a
```

##### tratamiento

```python
universidades = universidades[universidades['Comunidad autónoma']!='Estado']
```

```python
universidades['municipio_limpio'] = universidades['Municipio  de la unidad'].apply(doClean)
```

```python
universidades['ccaa'] = universidades['Comunidad autónoma'].apply(doClean)
```

```python
universidades['num_titulaciones'] = universidades['Número de titulaciones (Ciclo, Grado,  Máster y Doctorado)']
```

```python
universidades_group = universidades.groupby(by=['Universidad', 'municipio_limpio', 'ccaa', 'Tipo de universidad'], as_index=False)['num_titulaciones'].sum() 
```

```python
universidades_group = universidades_group.groupby(by=['municipio_limpio', 'ccaa', 'Tipo de universidad'], as_index=False).agg({'num_titulaciones':'mean', 'Universidad':'count'})
```

```python
universidades_group.head()
```

```python
universidades_group = universidades_group.pivot_table(index=['municipio_limpio', 'ccaa'], columns='Tipo de universidad',  values=['num_titulaciones', 'Universidad'], aggfunc='sum').reset_index()
```

```python
universidades_group.head()
```

```python
universidades_group.columns = ['municipio_limpio', 'ccaa' ,'num_universidades_privadas', 'num_universidades_publicas' , 'num_tit_uni_privadas', 'num_tit_uni_publicas']
```
