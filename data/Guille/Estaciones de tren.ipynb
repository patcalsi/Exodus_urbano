```python
def doClean(object):   
    
    # Quitar las expresiones tras la barra /
    a = re.sub('\/.*', '', object)
    # Quitar las aclaraciones entre paréntesis
    a = re.sub('\(.*\)', '', a)
    # Quitar acentos 
    a = unidecode.unidecode(a)
    # Quitar aquello que no sean palabras o cosa que se le parezca
    a = re.sub('\W', ' ', a) 
    # Quitar espacios extra en caso de haber
    a = re.sub('\s+', ' ', a)
    # Pasar el texto a minúsulas
    a = a.lower() 
    a = re.sub(' ', '_', a)
    #a = a.strip() 
    
    return a
```

```python
data['POBLACION'] = data['POBLACION'].apply(doClean)
data['PROVINCIA'] = data['PROVINCIA'].apply(doClean)
data
```

```python
data.replace(to_replace="Si",value="SI",inplace=True)
data.replace(to_replace="No",value="NO",inplace=True)
data
```

```python
count = pd.DataFrame(data['POBLACION'].value_counts().reset_index().values,columns=['poblacion','numero_estaciones'])
countindex = count.sort_index(axis = 0, ascending=True)
countindex
```

```python
poblacion = data.POBLACION[data.AVE=='SI'].unique()
len(poblacion)
ave = []
for i in range (0,42):
    ave.append('SI')
ave

d = {'poblacion': poblacion, 'ave': ave}
estaciones_ave = pd.DataFrame(data = d)
estaciones_ave
```

```python
data_ave = pd.merge(countindex,estaciones_ave,left_on='poblacion',right_on='poblacion',how='left')
```

```python
poblacion_cercanias = data.POBLACION[data.CERCANIAS=='SI'].unique()
len(poblacion_cercanias)
cercanias = []
for i in range (0,len(poblacion_cercanias)):
    cercanias.append('SI')
cercanias

d_c = {'poblacion': poblacion_cercanias, 'cercanias': cercanias}
estaciones_cercanias = pd.DataFrame(data = d_c)
estaciones_cercanias
```

```python
data_ave_cercanias = pd.merge(data_ave,estaciones_cercanias,left_on='poblacion',right_on='poblacion',how='left')
data_ave_cercanias
```

```python
poblacion_feve = data.POBLACION[data.FEVE=='SI'].unique()
len(poblacion_feve)
feve = []
for i in range (0,len(poblacion_feve)):
    feve.append('SI')
feve

d_f = {'poblacion': poblacion_feve, 'feve': feve}
estaciones_feve = pd.DataFrame(data = d_f)
estaciones_feve
```

```python
data_ave_cercanias_feve = pd.merge(data_ave_cercanias,estaciones_feve,left_on='poblacion',right_on='poblacion',how='left')
data_ave_cercanias_feve
```

```python
final_data = data_ave_cercanias_feve.fillna('NO')
final_data
```
