# Función que calcula distancia entre dos puntos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import unidecode
import os
import math
from math import * #sin,atan2,radians,cos
```

```python
#distancia en km partiendo de coordendadas
def distance(x_latitude1,y_longitude1,x_latitude2,y_longitude2):
    R = 6373.0
    a = 0
    c = 0

    a = sin(abs(radians(x_latitude1)-radians(x_latitude2))/ 2)**2 + cos(radians(x_latitude1)) * cos(radians(x_latitude2)) * sin(abs(radians(y_longitude1)-radians(y_longitude2))/ 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c  
```

```python
def standardize_headers(x):
    return x.replace(' ', '_').lower()
```

```python
def doClean(object):   
    
    # Quitar las expresiones tras la barra /
    a = re.sub('\/.*', '', object)
    # Quitar las aclaraciones entre paréntesis
    a = re.sub('\(.*\)', '', a)
    # Quitar acentos 
    a = unidecode.unidecode(a)
    # Quitar aquello que no sean palabras o cosa que se le parezca
    a = re.sub('\W', ' ', a) 
    # Quitar espacios extra en caso de haber
    a = re.sub('\s+', ' ', a)
    # Pasar el texto a minúsulas
    a = a.lower() 
    a = re.sub(' ', '_', a)
    #a = a.strip() 
    
    return a
```

```python
latitud_longitud_altitud=pd.read_csv('listado-longitud-latitud-municipios-espana.csv',sep = ';')
```

```python
latitud_longitud_altitud_clean=latitud_longitud_altitud.rename(columns=standardize_headers)
```

```python
latitud_longitud_altitud_clean.drop(['unnamed:_6','unnamed:_7','unnamed:_8'], axis=1,  inplace=True)
```

```python
latitud_longitud_altitud_clean.comunidad=latitud_longitud_altitud_clean.comunidad.apply(doClean)
latitud_longitud_altitud_clean.provincia=latitud_longitud_altitud_clean.provincia.apply(doClean)
latitud_longitud_altitud_clean['población']=latitud_longitud_altitud_clean['población'].apply(doClean)
```
