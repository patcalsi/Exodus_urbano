##### Relleno de nulls en el campo Cities

```python
df.fillna('', inplace=True)
```

```python
def doClean(text):   
    
    # Quitar las expresiones tras la barra /
    a = re.sub('\/.*', '', text)
    # Quitar las aclaraciones entre paréntesis
    a = re.sub('\(.*\)', '', a)
    # Quitar acentos 
    a = unidecode.unidecode(a)
    # Quitar aquello que no sean palabras o cosa que se le parezca
    a = re.sub('\W', ' ', a) 
    # Quitar espacios extra en caso de haber
    a = re.sub('\s+', ' ', a)
    # Pasar el texto a minúsulas
    a = a.lower() 
    a = re.sub(' ', '_', a)
    #a = a.strip() 
    
    return a
```

##### Limpieza campos categoricos

```python
df['municipio'] = df['Cities'].apply(doClean)
```

```python
df.head()
```

```python
df['provincia'] = df['Province'].apply(doClean)
```

```python
df['ccaa'] = df['Autonomous Community'].apply(doClean)
```

```python
df.head()
```

```python
df['cp'] = df['Postal code'].apply("{:05d}".format)
```

##### Eliminación de duplicados

```python
df[df.duplicated(subset=['cp', 'municipio', 'provincia', 'ccaa'])]
```

```python
maestro_municipios = df[['cp', 'municipio', 'provincia', 'ccaa']].drop_duplicates()
```
