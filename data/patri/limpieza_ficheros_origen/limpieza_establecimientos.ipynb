```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import unidecode
import os
```

#### Funciones

```python
def doClean(a):   
    a = str(a)
    # Quitar las expresiones tras la barra /
    #a = re.sub('\/.*', '', a)
    # Quitar las aclaraciones entre paréntesis
    a = re.sub('\(.*\)', '', a)
    # Quitar acentos 
    a = unidecode.unidecode(a)
    # Quitar aquello que no sean palabras o cosa que se le parezca
    a = re.sub('\W', ' ', a) 
    # Quitar espacios extra en caso de haber
    a = re.sub('\s+', ' ', a)
    # Pasar el texto a minúsulas
    a = a.lower() 
    # Quitar espacio izq
    a = a.lstrip()
    # Quitar espacio derecha
    a = a.rstrip()
    a = re.sub(' ', '_', a)
    #a = a.strip() 
    
    return a
```

##### tratamiento

```python
establecimientos['municipio_limpio'] = establecimientos['Localidad'].apply(doClean)
```

```python
establecimientos['ccaa'] = establecimientos['Comunidad autónoma'].apply(doClean)
```

```python
establecimientos['codigo_postal'] = establecimientos['Cód postal']
```

```python
establecimientos[establecimientos['municipio_limpio']=='segorbe'].head(2)
```

```python
establecimientos = establecimientos.groupby(by=['codigo_postal', 'municipio_limpio', 'ccaa', 'Oficio'], as_index=False)['Establecimiento'].count() 
```

```python
establecimientos.head(5)
```

```python
establecimientos = establecimientos.pivot_table(index=['codigo_postal', 'municipio_limpio', 'ccaa'], columns='Oficio',  values=['Establecimiento'], aggfunc='sum').reset_index()
```
